# Part 1: Using Matplotlib To

This is how to set up matplotlib

```python
plt.figure(figsize=(10,6))    #(figsize=(20,15)
```

### Part 1.1

```python
print(concatenated_all_df["country"].unique())
```

```python
selected_countries = ['afg', 'ind', 'bol', 'aus', 'chn']

#Calulating the list total population by country for each year
grouped_df = concatenated_all_df.groupby(['country', 'year'])['population'].sum().reset_index()

line_styles = ['-', '--', '-.', ':', '-']
colors = ['blue', 'green', 'red', 'purple', 'orange']

for i, country in enumerate(selected_countries):
    country_data = grouped_df[grouped_df['country'] == country]
    plt.plot(country_data['year'], 
             country_data['population'], 
             label=country,
             linestyle = line_styles[i])
        
             
    
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.title('Population Trends Over Time by Country')
plt.legend(title="Country")
plt.grid(True)
plt.show()
```

### Part 1.2

```python
# Creating age groups and filtering DataFrame
labels = ['0-14', '15-30', '31-45', '46-65', '66+']
bins = [0, 14, 30, 45, 65, 100]  # Set `100` to the max age if it's different

concatenated_all_df['age_group'] = pd.cut(concatenated_all_df['age'], bins=bins, labels=labels, right=True)

years_of_interest = [1950, 1985, 2020]
filtered_df = concatenated_all_df[concatenated_all_df['year'].isin(years_of_interest)]

# Grouping and calculating percentages
grouped_df = filtered_df.groupby(['year', 'age_group']).size().unstack(fill_value=0)
percentage_df = grouped_df.div(grouped_df.sum(axis=1), axis=0) * 100
percentage_df = percentage_df.fillna(0)  # Fill NaN values if any

# Plotting
colors = ['blue', 'green', 'red','purple']
ax = percentage_df.plot(kind='bar', stacked=True, figsize=(10, 6), color=colors)

# Adding percentage annotations
for i, year in enumerate(percentage_df.index):
    cumulative = 0
    for j, age_group in enumerate(percentage_df.columns):
        value = percentage_df.loc[year, age_group]
        if value > 0:  # Only annotate non-zero values
            cumulative += value / 2  # Get halfway point in stacked segment
            ax.text(
                i, cumulative, f"{value:.1f}%",
                ha='center', va='center', fontsize=8, color='black'
            )
            cumulative += value / 2  # Move to the top of this segment for the next one

# Adding labels and title
plt.title('Age Distribution by Year (Percentage)')
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.legend(title='Age Groups')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

### 1.2.1

```python
specified_year = 2000

filtered_df = concatenated_all_df[(concatenated_all_df['year'] == specified_year) & 
                                   (concatenated_all_df['country'].isin(selected_countries))]

filtered_df['age_group'] = pd.cut(filtered_df['age'], bins=bins, labels=labels, right=True)

grouped_df = filtered_df.groupby(['country', 'age_group'])['population'].sum().unstack(fill_value=0)

plt.figure(figsize=(10, 6))
sns.heatmap(grouped_df, annot=True, fmt="d", cmap="YlGnBu", cbar_kws={'label': 'Population'})

plt.title(f'Population Distribution Across Age Groups in {specified_year}')
plt.xlabel('Age Groups')
plt.ylabel('Countries')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

### 1.2.2

```python
country_1 = selected_countries[0]
country_2 = selected_countries[1]

filtered_df = concatenated_all_df[(concatenated_all_df['year'] == specified_year) & 
                                   (concatenated_all_df['country'].isin([country_1, country_2]))]

plt.figure(figsize=(12, 6))
sns.violinplot(x='country', y='age', hue='gender', data=filtered_df, split=True, inner="quartile", palette="muted")

plt.title(f'Age Distribution Comparison in {specified_year} Between {country_1} and {country_2}')
plt.xlabel('Countries')
plt.ylabel('Age')
plt.legend(title='Gender')
plt.grid(True)

plt.show()
```

# Part 2: Create a pairplot comparing population metrics across four countries:
1. Select four countries of your choice
2. Include all variables in the dataset
3. Use the country as the hue parameter to distinguish between countries
4. Write a brief summary of any patterns or relationships revealed by the pairplot

```python
# Selecting the first four countries
new_selected_countries = selected_countries[0:4]

# Filtering the dataframe to only include the selected countries
filtered_df = concatenated_all_df[concatenated_all_df['country'].isin(new_selected_countries)]

# Ensure `filtered_df` has only the relevant numeric columns for population metrics
# (e.g., 'population', 'year', etc. - adjust based on actual columns in `filtered_df`)
filtered_df = filtered_df[['country', 'year', 'population']]  # Add other numeric columns if needed

# Plotting the pair plot
sns.pairplot(filtered_df, hue='country', markers=["o", "s", "D", "P"], palette="muted")

# Adjusting the title for better positioning
plt.suptitle('Pairplot Comparing Population Metrics Across Selected Countries', y=1.05)
plt.show()
```
